```python
ds1 = dcmread(destination_path+"/"+images_list[0], stop_before_pixels=True)

def dictify(ds1):
    output = dict()
    for elem in ds1:
        if elem.VR != 'SQ': 
            output[elem.tag] = elem.value
        else:
            output[elem.tag] = [dictify(item) for item in elem]
    return output

train_metadata = []
for i in images_list:
    train_metadata.append(dictify(ds1))
        
```
